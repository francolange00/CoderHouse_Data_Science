En esta primera entrega se presentan 3 datasets potenciales para llevar a cabo el proyecto final del curso de Data Science:

Los 3 datasets fueron obtenidos de la página de Kaggle (www.kaggle.com) de los siguientes links:

-Leo Satellite Gloabl fire dataset: https://www.kaggle.com/datasets/brsdincer/leo-satellite-global-fire-data-1-year-nasa
-World cup dataset: https://www.kaggle.com/datasets/joebeachcapital/fifa-world-cups
-Earthquakes dataset: https://www.kaggle.com/datasets/usgs/earthquake-database



```python
import pandas as pd

#A continuación se crean las variables que guardarán la dirección de los datasets. 
#Los mismos se encuentran almacenados en Github

url_Global_fire = 'https://raw.githubusercontent.com/francolange00/CoderHouse_Data_Science/main/Leo_Satellite_Global%20_Fire-Dataset.csv'
url_Worldcup = 'https://raw.githubusercontent.com/francolange00/CoderHouse_Data_Science/main/world_cup_dataset.csv'
url_Earthquakes = 'https://raw.githubusercontent.com/francolange00/CoderHouse_Data_Science/main/Earthquakes-Dataset.csv'

#Se crean los dataframes para cada dataset:
df_Global_fire = pd.read_csv(url_Global_fire, index_col=0)
df_Worldcup = pd.read_csv(url_Worldcup, index_col=0)
df_Earthquakes = pd.read_csv(url_Earthquakes, index_col=0)

#Se muestran las dimensiones de cada dataframe:
rows_df_Global_fire, columns_df_Global_fire = df_Global_fire.shape

print(f'El Dataset "Global Fire" tiene {rows_df_Global_fire} filas y {columns_df_Global_fire} columnas.')

rows_df_Worldcup, columns_df_Worldcup = df_Worldcup.shape

print(f'El Dataset "Worldcup" tiene {rows_df_Worldcup} filas y {columns_df_Worldcup} columnas.')

rows_df_Earthquakes, columns_df_Earthquakes = df_Earthquakes.shape

print(f'El Dataset "Earthquakes" tiene {rows_df_Earthquakes} filas y {columns_df_Earthquakes} columnas.')

```

Teniendo en cuenta estos 3 datasets, se destacan las siguientes posibles variables a analizar para cada caso:

-Leo Satellite Gloabl fire dataset:
    -FRP: Representa la potencia radiativa del fuego integrada en el píxel en MW (megavatios). 
    -Fire_Type: Representa el punto caliente inferido (incendio de vegetacion, volcán activo, otra fuente terreste estática o                 fuera de costa
    -Bright_ti4: Temperatura de brillo en el canal I-4 del sensor VIIRS, medida en Kelvin
 
-World cup dataset:
    -Goals: Goles efectuados
    -Passes: Pases concretados
    -Fouls: Faltas cometidas
    -Fouls_r: Faltas recibidas
    -Assists: Asistencias realizadas
   
-Earthquakes dataset:
    -Lat: Latitud
    -Long: Longitud
    -Magnitude: Magnitud del terremoto
    -Status: Estado del terremoto
    -Depth: Profundidad del terremoto
