```python
Pregunta_4 = """Respondiendo a la última pregunta planteada al inicio: ***¿La cantidad de incendios es constante estacionalmente, o se tienen ciertas difernecias en función de la estaciones del año?***

En este caso podemos reutilizar una gráfica obtenida en la entrega anterior en donde se muestra la sumatoria de incendios mes a mes a lo largo de todo un año:
"""

display(Markdown(Pregunta_4))
```

```python
import calendar

df_Global_fire['acq_date'] = pd.to_datetime(df_Global_fire['acq_date'])

df_Global_fire['month'] = df_Global_fire['acq_date'].dt.month
df_Global_fire['year'] = df_Global_fire['acq_date'].dt.year

#Se agrupa y cuentan los incendios por año y mes
Fires_sum = df_Global_fire.groupby(['year', 'month']).size().reset_index(name='count')

#Se mapea los números de mes a nombres de mes
Fires_sum['month'] = Fires_sum['month'].apply(lambda x: calendar.month_abbr[x])

#Se Crea una gráfica de barras
plt.figure(figsize=(12, 6))
plt.bar(Fires_sum['year'].astype(str) + '-' + Fires_sum['month'], Fires_sum['count'], color='r')
plt.ylabel('Suma total')
plt.title('Numero de incendios en 2021 (Turkey)')
plt.xticks(rotation=45)

# Se muestra la gráfica
plt.tight_layout()
plt.show()
```

```python
Respuesta_4 = """Como era de esperarse, tenemos un gran incremento en el número de incedios en la estación de verano principalmente. Por lo que se responde a la pregunta de que dichos incendios no son constantes a lo largo del año y por lo tanto a hay que poner foco entre los meses de junio y agosto que es en donde más casos de incendios se registran. """

display(Markdown(Respuesta_4))
```

A continuación se desarrollará un modelo de regresión lineal con el fin de poder predecir el parámetro "brigthness" del dataset, que nos ayudará a saber el grado de intensidad de un determinado incendio. Esto es potencialmente util para tener una idea del grado de peligrosidad de un nuevo foco de incendio y poder actuar de la mejor forma posible, optimizando los recursos disponibles.  
 
Lo primero que se hizo fue definir nuestra variable target, en este caso "brigthness, y las "features" que en una primera iteración se tomaron: "latitude", "longitude", "scan", "track" y "confidence". Se dejaron fuera del modelo todos aquellos parámetros que no eran numéricos o que se calculan a raíz del parámetro "brigthness". Luego creó un modelo de regresión lineal utilizando la función "LinearRegression" de la libreria "sklearn", previamente dividiendo el dataset destinando un 80% para entrenarlo y un 20% 
    para testearlo.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

target = df_Global_fire['brightness']

features = df_Global_fire[["latitude", "longitude", "scan", "track", "confidence"]]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
# Se crea el modelo
lr = LinearRegression()
# Se ajustar el modelo con X_train y y_train
lr.fit(X_train,y_train)
# Se pedice con X_test
y_pred = lr.predict(X_test)

```


A continuación, para tener un punto de partida, se obtienen distintas métricas del redimiento de nuestro modelo, el cual fue entrenado utilizando todos los parámetros posibles previamente mencionados:


```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

print("MAE ",round(mean_absolute_error(y_test,y_pred),4))

import numpy as np
print("RMSE",round(np.log(np.sqrt(mean_squared_error(y_test,y_pred))),4))

from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print("R2  ",round(r2,4))
```

Ahora, partiendo de estos números y utilizando el método de feature selection "forward selection" se busca el número minimo de features posibles que permiten obtener valores similares o incluso mejores que nuestro modelo base:

```python
import sys
import joblib
sys.modules['sklearn.externals.joblib'] = joblib

#Librerias
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.linear_model import LinearRegression
# Sequential Forward Selection(sfs)
sfs = SFS(LinearRegression(),
          k_features=4,
          forward=True,
          floating=False,
          scoring = 'r2',
          cv = 0)
sfs.fit(features, target)
print("Lista de features seleccionados: ",sfs.k_feature_names_)     #Lista final de features

features = df_Global_fire.loc[:, sfs.k_feature_names_]


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.4, random_state=42)
# Se crea el modelo
lr = LinearRegression()
# Ajustar el modelo con X_train y y_train
lr.fit(X_train,y_train)
# PRedecir con X_test
y_pred_train =lr.predict(X_train)

y_pred = lr.predict(X_test)


print("MAE ",round(mean_absolute_error(y_test,y_pred),4))


print("RMSE",round(np.log(np.sqrt(mean_squared_error(y_test,y_pred))),4))

r2 = r2_score(y_test,y_pred)
print("R2  ",round(r2,4))



```

Luego probar diferentes números de features posibles, se obtuvo que el mejor resultado(incluso mejor que utilizando todos los parámetros) lo obtenemos seleccionando la latitud, longitud y el nivel de confianza de los datos. Por otro lado, si bien el nivel de confianza nos brinda información sobre que tan precisos y confiables son los datos aportados por el satélite, y además que en el análisis anterior figura dentro de la lista de features con las que obtuvimos mejores resultados, se entiende que este parámetro no tiene relación con la el nivel de intencidad de los incendios, por lo que idealmente no se debería de utilizar como feature y quedarnos solo con la latitud y longitud. 
Por otro lado, tambien se probó de utilizar diferentes porcentajes de segmentación para los datos de test y entrenamiento, resultado un 80% para entrenamiento y un 20% para test como la forma más optima para este dataset.

Otro punto importante a tener en cuenta es que según el MAE(error absoluto medio), la diferencia absoluta entre el valor objetivo y el valor predicho por el modelo es de 12,4634, y teniendo en cuenta que la mediana de este parámetro(319.7) en el dataset, nos da como resultado una variación aproximada del 3,89% si tomamos como referencia este valor de mediana. Es por esto que a grandes rasgos podemos considerar que los valores predichos por el modelo son bastante cercanos.

Para poder determinar si nuestro modelo podría llegar a sufrir las causas de overfitting o underfitting se analizó primero los datos obtenidos anteriormente, enfocandonos en el error abosulo medio como métrica principal. Teniendo esto en cuenta, se descartó casi de inmediato la posibilidad de tener underfitting, debido a que el valor de MAE era relativamente aceptable comparado con la mediana del parámetro bajo análisis. Por lo que faltaba analizar si nuestro modelo era o no candidatos a sufrir un overfitting. En este caso lo que se hizo fue comparar el valor de MAE obtenido con los datos de entrenamiento versus el MAE obtenido con los datos de test, si el MAE obtenido pra los valores de entrenamiento es considerablemente mejor, podemos estar seguros de que nuestro modelo no fue capaz de generalizar los datos y por lo tanto fue victima de un overfitting.

 

```python
y_pred_train =lr.predict(X_train)
mae = mean_absolute_error(y_train, y_pred_train)
print(f"Error absoluto medio (MAE) en datos de entrenamiento: {mae:.4f}")
```
